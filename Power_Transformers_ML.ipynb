```python
import numpy as np, pandas as pd, seaborn as sns

log_normal = np.random.lognormal(2,1,1000)
exponential = np.random.exponential(1,1000)
pareto = np.random.pareto(3,1000)
power = np.random.power(5,1000)
uniform = np.random.uniform(-1,0,1000)
df = pd.DataFrame({"log_normal":log_normal,"exponential":exponential,"pareto":pareto,"power":power,"uniform":uniform})
```

```python
import sklearn
from sklearn.model_selection import train_test_split
dfTrain,dfTest= train_test_split (df, test_size=0.20,random_state=32)
```

```python
from sklearn.preprocessing import PowerTransformer

pt=PowerTransformer()

YeoDF1= pd.DataFrame(pt.fit_transform(dfTrain), columns=dfTrain.columns, index=dfTrain.index)
```
